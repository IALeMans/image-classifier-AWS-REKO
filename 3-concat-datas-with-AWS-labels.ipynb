```python
import os
import pandas as pd
```

```python
def get_files_list(datadir):
    all_files = os.listdir(os.path.abspath(datadir))
    return list(filter(lambda file: file.endswith('.csv'), all_files))
```

```python
datadirtemp = "./collect_temp2/"
file_out = "Mairie_appels_filename_labels2.csv"

list_csv = get_files_list(datadirtemp)
```

```python
df = pd.DataFrame()
for i, val in enumerate(list_csv):
    print(f'proceeding {val}')
    df_temp = pd.read_csv(datadirtemp+val,sep=';') 
    df = pd.concat([df, df_temp],sort=False)
```

```python
df = df.set_index('cle').fillna(0).drop(['Unnamed: 0'],axis=1)
```
